```python
# Charge les données Orphanet depuis un fichier CSV dans un DataFrame pour les exploiter facilement. 
# `read_csv()` permet de lire le fichier CSV.
import pandas as pd
df_orphanet = pd.read_csv('data_orphanet.csv')
```

```python
# Affiche le nombre de lignes et de colonnes du DataFrame avec `shape`.
print(df_orphanet.shape)
```

```python
#Affiche le type de données de chaque colonne du DataFrame à l’aide de `dtypes`.
print(df_orphanet.dtypes) 
```

```python
# Compte le nombre de valeurs manquantes par colonne avec `isnull().sum()`.
print(df_orphanet.isnull().sum()) 
```

```python
# Compte le nombre de lignes dupliquées dans le DataFrame avec `duplicated().sum()`.
print(df_orphanet.duplicated().sum()) 
```

```python
# Affiche la liste des noms de colonnes du DataFrame avec `columns`.
print(df_orphanet.columns)
```

```python
# Affiche les noms des colonnes du DataFrame sous forme de liste avec `list(df_orphanet.columns)`.
print(list(df_orphanet.columns))
```

```python
# Supprime les espaces inutiles dans les noms de colonnes pour éviter les erreurs d’accès. 
# `str.strip()` nettoie les intitulés de colonnes.
df_orphanet.columns = df_orphanet.columns.str.strip()
```

```python
# Supprime les doublons en se basant sur la colonne `orphacode` pour garder une seule ligne par maladie.
df_orphanet = df_orphanet.drop_duplicates(subset=['orphacode'])
```

```python
# Nettoie et uniformise les noms de maladies en supprimant les espaces et en mettant une casse 
# standard avec `str.strip()` et `str.title()`.
df_orphanet['maladie'] = df_orphanet['maladie'].str.strip().str.title() 
```

```python
# Sépare la colonne `source` en liste à partir du séparateur `;` pour structurer les différentes sources.
df_orphanet['source'] = df_orphanet['source'].str.split('; ') 
```

```python
# Nettoie chaque élément de la colonne `source` en supprimant les caractères spéciaux et en normalisant le texte en minuscules. 
# `re.sub()` permet de filtrer les caractères indésirables.
import re
df_orphanet['source'] = df_orphanet['source'].apply( 
    lambda terms: [re.sub(r'[^a-z0-9 ]', '', t.lower().strip()) for t in terms] 
) 
```
