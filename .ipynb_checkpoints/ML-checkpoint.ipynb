```python
# Chargement du dataset unifié (étape 2) — génération automatique si absent
import os

if os.path.exists('dataset_unifie.csv'):
    df = pd.read_csv('dataset_unifie.csv')
    if 'hpo_clean' in df.columns and isinstance(df['hpo_clean'].iloc[0], str):
        df['hpo_clean'] = df['hpo_clean'].apply(
            lambda x: ast.literal_eval(x) if x.startswith('[') else x.split('; ')
        )
    print(f'✅ Dataset chargé : {df.shape}')
else:
    # Génération d'un dataset simulé complet avec toutes les colonnes requises
    print('⚠️  dataset_unifie.csv absent — génération simulation...')
    N = 1580
    categories = ['musculaire','neurologique','osseux','articulaire','vertébral','autre']
    cat_w      = [0.28, 0.22, 0.18, 0.16, 0.10, 0.06]
    spec_map   = {'musculaire':'Rhumatologie','neurologique':'Neurologie',
                  'osseux':'Orthopédie','articulaire':'Rhumatologie',
                  'vertébral':'Neurochirurgie','autre':'Médecine générale'}
    hpo_bank   = {
        'musculaire':   ['muscle weakness','myalgia','muscle atrophy','cramp',
                         'fatigue','proximal muscle weakness','myopathy','exercise intolerance'],
        'neurologique': ['peripheral neuropathy','pain','sensory loss','tremor',
                         'paralysis','nerve conduction','weakness','paresthesia','spasticity'],
        'osseux':       ['bone pain','fracture','osteoporosis','bone deformity',
                         'short stature','skeletal dysplasia','osteolysis','joint hypermobility'],
        'articulaire':  ['joint pain','arthralgia','joint swelling','arthritis',
                         'stiffness','limited range of motion','synovitis','joint deformity'],
        'vertébral':    ['back pain','scoliosis','spinal stenosis','radiculopathy',
                         'kyphosis','neck pain','disc herniation','vertebral instability'],
        'autre':        ['pain','fatigue','disability','inflammation','chronic pain','weakness']
    }
    zones_map  = {
        'musculaire':['upper limb','lower limb','trunk','shoulder'],
        'neurologique':['lower limb','spine','upper limb','head'],
        'osseux':['lower limb','spine','upper limb','pelvis'],
        'articulaire':['knee','hip','shoulder','wrist','ankle'],
        'vertébral':['lumbar','cervical','thoracic','sacral'],
        'autre':['whole body','trunk','head']
    }
    rows = []
    for i in range(N):
        cat   = np.random.choice(categories, p=cat_w)
        n_s   = max(1, int(np.random.lognormal(2.1, 0.6)))
        hpo   = list(np.random.choice(hpo_bank[cat], size=min(n_s, len(hpo_bank[cat])), replace=False))
        sev_k = ['pain','weakness','paralysis','disability','deformity','atrophy']
        sev   = sum(1 for t in hpo if any(k in t for k in sev_k))
        rows.append({
            'disease_name':         f'Disease_{cat}_{i}',
            'orpha_code':            1000+i,
            'num_symptoms_clean':    len(hpo),
            'hpo_clean':             hpo,
            'anatomical_category':   cat,
            'body_zone':             np.random.choice(zones_map[cat]),
            'severity_score':        sev,
            'medical_specialty':     spec_map[cat],
            'age_onset':             round(np.clip(np.random.normal(38,18),5,85),1),
            'pain_intensity':        round(np.clip(np.random.normal(5.5+sev*0.4,1.8),1,10),1),
            'global_prevalence_median': np.random.exponential(15),
        })
    df = pd.DataFrame(rows)
    print(f'✅ Dataset simulé : {df.shape}')

df.head(3)
```

## 4.1 — Définition du problème ML & Feature Engineering

#### * Définit les variables d’entrée et la cible du modèle de machine learning.
#### * Précise qu’il s’agit d’un problème de classification multi-classes.
#### * Présente la stratégie de séparation des données et de validation croisée.
#### * Affiche la répartition des spécialités médicales à prédire.



```python
# Définition claire du problème : ce que le modèle prend en entrée et ce qu'il prédit
print('╔══════════════════════════════════════════════════════════╗')
print('║           PROBLÈME ML — elaraMed                        ║')
print('╠══════════════════════════════════════════════════════════╣')
print('║  INPUT   → zone corporelle + symptômes HPO              ║')
print('║            + âge + intensité douleur + sévérité         ║')
print('╠══════════════════════════════════════════════════════════╣')
print('║  TARGET  → spécialité médicale recommandée              ║')
print('║  (cible secondaire optionnelle : catégorie anatomique)   ║')
print('╠══════════════════════════════════════════════════════════╣')
print('║  TYPE    → Classification multi-classes                 ║')
print('║  SPLIT   → 80% train / 20% test  (stratifié)            ║')
print('║  CV      → StratifiedKFold  k = 5                       ║')
print('╚══════════════════════════════════════════════════════════╝')
print()
print('Classes cibles (medical_specialty) :')
print(df['medical_specialty'].value_counts().to_string())
```

### Prépare les données pour l’apprentissage automatique en transformant les symptômes, les variables cliniques et les zones anatomiques en caractéristiques exploitables par l’IA.
### Encode la spécialité médicale cible sous forme numérique afin de permettre au modèle de classification de prédire automatiquement le spécialiste le plus adapté à partir du profil du patient.


```python
# Construction de la matrice de features X et de la cible y
# Les symptômes HPO sont convertis en vecteur TF-IDF ; les features numériques sont ajoutées séparément

# --- Feature 1 : texte HPO → chaîne jointe pour TF-IDF ---
df['hpo_text'] = df['hpo_clean'].apply(
    lambda x: ' '.join(x) if isinstance(x, list) else str(x)
)

# --- Feature 2 : encodage one-hot de la zone corporelle ---
df_zone = pd.get_dummies(df['body_zone'], prefix='zone')

# --- Feature 3 : variables numériques normalisées ---
num_feats = ['num_symptoms_clean','severity_score','age_onset','pain_intensity']
for col in num_feats:
    if col not in df.columns:
        df[col] = 0
df_num = df[num_feats].fillna(df[num_feats].median())
# Normalisation min-max manuelle pour garder la compatibilité avec MultinomialNB (valeurs ≥ 0)
df_num_norm = (df_num - df_num.min()) / (df_num.max() - df_num.min() + 1e-9)

# --- Cible : spécialité médicale encodée en entier ---
le = LabelEncoder()
y  = le.fit_transform(df['medical_specialty'])

print(f'Classes encodées : {dict(enumerate(le.classes_))}')
print(f'Distribution cible :')
print(pd.Series(y).value_counts().rename(index=dict(enumerate(le.classes_))).to_string())
```

#### Transforme les symptômes HPO en vecteurs numériques TF-IDF afin que l’IA puisse analyser l’information textuelle.
#### Fusionne les données textuelles, cliniques et anatomiques dans une matrice de caractéristiques, puis sépare les données en ensembles d’entraînement et de test pour évaluer les performances du modèle.


```python
# Vectorisation TF-IDF des symptômes HPO et concaténation avec les features numériques et one-hot
# Le résultat est une matrice sparse compatible avec tous les modèles sklearn

tfidf = TfidfVectorizer(max_features=300, ngram_range=(1,2), min_df=2)
X_tfidf = tfidf.fit_transform(df['hpo_text'])       # shape : (N, 300)

X_zone  = csr_matrix(df_zone.values.astype(float))  # shape : (N, nb_zones)
X_num   = csr_matrix(df_num_norm.values)             # shape : (N, 4)

# Concaténation horizontale de toutes les features en une seule matrice sparse
X = hstack([X_tfidf, X_zone, X_num])
print(f'Matrice X finale : {X.shape}  → {X.shape[1]} features par patient')

# Split train / test stratifié (80 / 20)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42, stratify=y
)
print(f'Train : {X_train.shape[0]} samples  |  Test : {X_test.shape[0]} samples')
```

## 4.2 — Entraînement des modèles

```python
# Définition des modèles de machine learning utilisés pour la comparaison.
# Chaque algorithme applique une approche différente pour prédire la spécialité médicale.

models = {
    'Random Forest': RandomForestClassifier(
        n_estimators=200,
        max_depth=None,
        min_samples_leaf=2,
        n_jobs=-1,
        random_state=42,
        class_weight='balanced'   # Compense le déséquilibre entre les classes.
    ),

    'Naive Bayes (Complement)': ComplementNB(
        alpha=0.5                 # Optimisé pour les données textuelles et les classes déséquilibrées.
    ),

    'KNN': KNeighborsClassifier(
        n_neighbors=7,
        weights='distance',       # Donne plus d'importance aux voisins les plus proches.
        metric='cosine',          # Mesure adaptée aux vecteurs TF-IDF.
        n_jobs=-1
    ),

    'XGBoost': xgb.XGBClassifier(
        n_estimators=300,
        max_depth=6,
        learning_rate=0.1,
        subsample=0.8,
        colsample_bytree=0.8,
        use_label_encoder=False,
        eval_metric='mlogloss',
        random_state=42,
        n_jobs=-1
    )
}

# Affichage des modèles retenus pour l'évaluation comparative.
print('Modèles définis :')
for name in models:
    print(f'  • {name}')
```

```python
# Entraîne et évalue chaque modèle afin de comparer leurs performances
# sur la recommandation automatique de spécialités médicales.

import time

results   = {}   # Stocke les métriques d'évaluation.
trained   = {}   # Conserve les modèles entraînés.
cv_scores = {}   # Stocke les résultats de validation croisée.

# Validation croisée stratifiée pour préserver la distribution des classes.
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

for name, model in models.items():
    print(f'\n─── {name} ───')

    # Apprentissage du modèle sur les données d'entraînement.
    t0 = time.time()
    model.fit(X_train, y_train)
    train_time = time.time() - t0

    # Prédiction des spécialités médicales sur l'ensemble de test.
    y_pred = model.predict(X_test)

    # Calcul des métriques de performance.
    acc    = accuracy_score(y_test, y_pred)
    f1_mac = f1_score(y_test, y_pred, average='macro')
    f1_wgt = f1_score(y_test, y_pred, average='weighted')

    # Évaluation de la robustesse du modèle par validation croisée.
    cv = cross_val_score(model, X_train, y_train, cv=skf,
                         scoring='accuracy', n_jobs=-1)

    # Sauvegarde des résultats obtenus.
    results[name] = {
        'accuracy': acc,
        'f1_macro': f1_mac,
        'f1_weighted': f1_wgt,
        'train_time': train_time
    }

    trained[name] = model
    cv_scores[name] = cv

    # Affichage des performances du modèle.
    print(f'  Accuracy     : {acc:.4f}')
    print(f'  F1 macro     : {f1_mac:.4f}')
    print(f'  F1 weighted  : {f1_wgt:.4f}')
    print(f'  CV (5-fold)  : {cv.mean():.4f} ± {cv.std():.4f}')
    print(f'  Temps entr.  : {train_time:.2f}s')

print('\n✅ Tous les modèles entraînés.')
```

## 4.3 — Évaluation détaillée

```python
# Génère un rapport détaillé de classification pour chaque modèle entraîné.
# Ce rapport permet d'évaluer les performances par classe (precision, recall, F1-score)
# et d'identifier les spécialités médicales les plus difficiles à prédire.

for name, model in trained.items():
    # Prédiction sur le jeu de test
    y_pred = model.predict(X_test)

    # Affichage du titre du modèle
    print(f"\n{'='*55}")
    print(f'  {name}')
    print('='*55)

    # Rapport de classification complet
    print(classification_report(
        y_test,
        y_pred,
        target_names=le.classes_,
        digits=3
    ))
```

```python
# Visualise les matrices de confusion des différents modèles afin d'analyser
# la qualité des prédictions et les erreurs de classification entre classes.

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten()

# Colormaps différents pour distinguer visuellement les modèles
cmaps_conf = ['Blues', 'Greens', 'Oranges', 'Purples']

for idx, (name, model) in enumerate(trained.items()):
    # Prédictions sur le jeu de test
    y_pred = model.predict(X_test)

    # Calcul de la matrice de confusion
    cm = confusion_matrix(y_test, y_pred)

    # Normalisation par ligne pour obtenir des proportions (0–1)
    cm_pct = cm.astype(float) / cm.sum(axis=1, keepdims=True)

    # Heatmap de la matrice normalisée
    sns.heatmap(
        cm_pct,
        annot=True,
        fmt='.2f',
        xticklabels=le.classes_,
        yticklabels=le.classes_,
        cmap=cmaps_conf[idx],
        ax=axes[idx],
        linewidths=0.5,
        linecolor='#E8E8E8',
        vmin=0,
        vmax=1,
        annot_kws={'size': 8}
    )

    # Affichage de l’accuracy dans le titre
    acc = results[name]['accuracy']
    axes[idx].set_title(
        f'{name}\nAccuracy = {acc:.3f}',
        fontsize=11,
        fontweight='bold',
        color=BLUE
    )

    axes[idx].set_xlabel('Prédit', fontsize=9)
    axes[idx].set_ylabel('Réel', fontsize=9)
    axes[idx].tick_params(axis='x', rotation=20, labelsize=8)
    axes[idx].tick_params(axis='y', rotation=0, labelsize=8)

# Titre global de la figure
plt.suptitle(
    'Matrices de confusion normalisées — 4 modèles',
    fontsize=14,
    fontweight='bold',
    color=BLUE,
    y=1.01
)

plt.tight_layout()
plt.savefig('fig_confusion_matrices.png', bbox_inches='tight')
plt.show()
```

```python
# Visualise la stabilité des modèles via les scores de validation croisée (k=5).
# Permet d'évaluer la robustesse et la variance des performances entre folds.

fig, ax = plt.subplots(figsize=(10, 5))

# Extraction des résultats de CV
cv_data  = list(cv_scores.values())
cv_names = list(cv_scores.keys())
colors_cv = PALETTE[:len(cv_names)]

# Boxplot des scores de cross-validation
bp = ax.boxplot(
    cv_data,
    patch_artist=True,
    widths=0.45,
    medianprops={'color': 'white', 'linewidth': 2.5}
)

# Coloration des boxplots par modèle
for patch, color in zip(bp['boxes'], colors_cv):
    patch.set_facecolor(color)
    patch.set_alpha(0.82)

# Stylisation des whiskers
for whisker in bp['whiskers']:
    whisker.set_linestyle('--')
    whisker.set_color('#888')

# Ajout des points individuels (chaque fold)
for i, (scores, color) in enumerate(zip(cv_data, colors_cv), start=1):
    jitter = np.random.uniform(-0.12, 0.12, size=len(scores))
    ax.scatter(
        np.full(len(scores), i) + jitter,
        scores,
        color=color,
        edgecolor='white',
        s=55,
        zorder=5,
        linewidth=0.8
    )

# Configuration des axes
ax.set_xticks(range(1, len(cv_names) + 1))
ax.set_xticklabels(cv_names, rotation=10, fontsize=10)
ax.set_ylabel('Accuracy (CV k=5)', fontsize=11)
ax.set_title(
    'Cross-validation — Stabilité des modèles (5 folds)',
    fontsize=13,
    fontweight='bold',
    color=BLUE
)

# Ajustement dynamique de l'échelle Y
ax.set_ylim(
    max(0, min(s.min() for s in cv_data) - 0.05),
    1.02
)

# Annotation des moyennes par modèle
for i, scores in enumerate(cv_data, start=1):
    ax.text(
        i,
        scores.max() + 0.012,
        f'μ={scores.mean():.3f}',
        ha='center',
        fontsize=9,
        color='#333'
    )

plt.tight_layout()
plt.savefig('fig_cross_validation.png', bbox_inches='tight')
plt.show()
```

```python
# Analyse l'importance des variables dans le modèle Random Forest
# afin d'interpréter les facteurs les plus influents dans la prédiction
# de la spécialité médicale.

rf_model = trained['Random Forest']

# Reconstruction des noms de features (TF-IDF + zones + variables numériques)
feature_names = (
    list(tfidf.get_feature_names_out()) +
    list(df_zone.columns) +
    num_feats
)

# Extraction des importances apprises par le modèle (critère de Gini)
importances = rf_model.feature_importances_

# Organisation des résultats dans un DataFrame trié
feat_df = pd.DataFrame({
    'feature': feature_names,
    'importance': importances
}).sort_values('importance', ascending=False).head(20)

# Visualisation des 20 features les plus influentes
fig, ax = plt.subplots(figsize=(10, 6))

# Codage couleur selon le type de feature
colors_imp = [
    BLUE if not f.startswith('zone_') and f not in num_feats
    else ('#F4A261' if f.startswith('zone_') else ACCENT)
    for f in feat_df['feature']
]

ax.barh(
    feat_df['feature'][::-1],
    feat_df['importance'][::-1],
    color=colors_imp[::-1],
    edgecolor='white',
    linewidth=0.5
)

# Légende explicative des types de features
from matplotlib.patches import Patch
legend_el = [
    Patch(color=BLUE, label='Terme HPO (TF-IDF)'),
    Patch(color='#F4A261', label='Zone corporelle (one-hot)'),
    Patch(color=ACCENT, label='Feature numérique'),
]

ax.legend(handles=legend_el, fontsize=9, loc='lower right')

# Titre et mise en forme
ax.set_title(
    'Importance des features — Random Forest (Top 20)',
    fontsize=13,
    fontweight='bold',
    color=BLUE
)

ax.set_xlabel('Importance (Gini)', fontsize=11)
ax.tick_params(axis='y', labelsize=9)

plt.tight_layout()
plt.savefig('fig_feature_importance.png', bbox_inches='tight')
plt.show()
```

## 4.4 — Comparaison finale des modèles

```python
# Compare les performances globales des modèles afin d'identifier
# celui offrant le meilleur compromis entre précision, robustesse et stabilité.

# Construction du tableau récapitulatif des résultats
results_df = pd.DataFrame(results).T.reset_index().rename(columns={'index': 'Modèle'})

# Tri selon la métrique principale (F1 weighted)
results_df = results_df.sort_values('f1_weighted', ascending=False).reset_index(drop=True)

# Ajout des statistiques de validation croisée
results_df['CV_mean'] = [cv_scores[m].mean() for m in results_df['Modèle']]
results_df['CV_std']  = [cv_scores[m].std()  for m in results_df['Modèle']]

# Arrondi des métriques pour une meilleure lisibilité
results_df[['accuracy','f1_macro','f1_weighted','CV_mean','CV_std']] = \
    results_df[['accuracy','f1_macro','f1_weighted','CV_mean','CV_std']].round(4)

# Formatage du temps d'entraînement
results_df['train_time'] = results_df['train_time'].round(2).astype(str) + 's'

# Affichage du tableau comparatif final
print('\n' + '='*70)
print('         COMPARAISON FINALE DES MODÈLES — elaraMed')
print('='*70)

print(
    results_df[[
        'Modèle','accuracy','f1_macro','f1_weighted',
        'CV_mean','CV_std','train_time'
    ]].to_string(index=False)
)

print('='*70)

# Sélection du meilleur modèle selon le F1-score pondéré
best = results_df.iloc[0]['Modèle']
print(f'\n🏆 Meilleur modèle (F1 weighted) : {best}')
```

```python
# Fonction de prédiction finale utilisant le meilleur modèle entraîné
# Simule un cas d'usage clinique : symptômes patient → recommandation de spécialité médicale

best_model = trained[best]

def predict_specialty(symptoms: list, body_zone: str,
                       age: float, pain: float, severity: int = 1) -> dict:
    """
    Prédit la spécialité médicale la plus probable pour un patient.

    Cette fonction transforme les entrées patient en vecteurs compatibles
    avec le pipeline ML (TF-IDF + one-hot + variables numériques).

    Args:
        symptoms  : liste de symptômes HPO (texte libre)
        body_zone : localisation anatomique
        age       : âge du patient
        pain      : intensité de la douleur (1–10)
        severity  : score clinique de sévérité

    Returns:
        dict : spécialité prédite + top-3 probabilités
    """

    # --- Transformation texte (symptômes → TF-IDF) ---
    hpo_str = ' '.join(symptoms)
    x_tfidf = tfidf.transform([hpo_str])

    # --- Encodage one-hot de la zone corporelle ---
    zone_row = {col: 0.0 for col in df_zone.columns}
    zone_key = f'zone_{body_zone}'
    if zone_key in zone_row:
        zone_row[zone_key] = 1.0
    x_zone = csr_matrix([[zone_row[c] for c in df_zone.columns]])

    # --- Normalisation des variables numériques ---
    raw_num = np.array([[len(symptoms), severity, age, pain]])
    mins = df_num.min().values
    maxs = df_num.max().values
    x_num_n = csr_matrix((raw_num - mins) / (maxs - mins + 1e-9))

    # --- Construction du vecteur final patient ---
    x_patient = hstack([x_tfidf, x_zone, x_num_n])

    # --- Prédiction de la classe ---
    pred_idx = best_model.predict(x_patient)[0]
    pred_spec = le.inverse_transform([pred_idx])[0]

    # --- Probabilités (si disponibles) ---
    if hasattr(best_model, 'predict_proba'):
        proba = best_model.predict_proba(x_patient)[0]
        top3 = sorted(zip(le.classes_, proba), key=lambda x: -x[1])[:3]
    else:
        top3 = [(pred_spec, 1.0)]

    return {
        'specialty': pred_spec,
        'top3': top3
    }


# ---------------- EXEMPLES D'UTILISATION ----------------

# Patient 1 : tableau articulaire
res1 = predict_specialty(
    symptoms=['joint pain', 'arthralgia', 'joint swelling', 'stiffness'],
    body_zone='knee',
    age=45,
    pain=7,
    severity=2
)

print(f'Patient 1 → Spécialité recommandée : {res1["specialty"]}')
print('  Top-3 :', [(s, f'{p:.2%}') for s, p in res1['top3']])

# Patient 2 : symptômes neurologiques
res2 = predict_specialty(
    symptoms=['peripheral neuropathy', 'sensory loss', 'weakness', 'paresthesia'],
    body_zone='lower limb',
    age=58,
    pain=5,
    severity=3
)

print(f'\nPatient 2 → Spécialité recommandée : {res2["specialty"]}')
print('  Top-3 :', [(s, f'{p:.2%}') for s, p in res2['top3']])

# Patient 3 : douleurs vertébrales
res3 = predict_specialty(
    symptoms=['back pain', 'scoliosis', 'radiculopathy', 'neck pain'],
    body_zone='lumbar',
    age=37,
    pain=8,
    severity=2
)

print(f'\nPatient 3 → Spécialité recommandée : {res3["specialty"]}')
print('  Top-3 :', [(s, f'{p:.2%}') for s, p in res3['top3']])
```
